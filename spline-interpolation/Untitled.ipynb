## Введение <a name='introduction'></a>

Интерполяцией называется метод нахождения неизвестных промежуточных значений некоторой функции по имеющемуся дискретному набору ее известных значений.

Сплайн-интерполяция - это вид интерполяции, которая выполняется при помощи сплайнов
 - Сплайн - это функция, область определения которой разбита на конечное число отрезков, на каждом из которых она совпадает с некоторым алгебраическим многочленом (полиномом). 
 
В данной лабораторной работе рассматривается интерполяция кубическими сплайнами, то есть полинома третьей степени.

## Алгоритм решения <a name='paragraph2'></a>

Пусть нам даны точки $[x_1, x_2, ..., x_n], [y_1, y_2, ..., y_n]$ тогда нам нужно на каждом промежутке $[x_i, x_{i+1}]$ определить кубический сплайн вида: $S_i(x)= a_i + b_i(x-x_i) + c_i(x-x_i)^2 + d_i(x-x_i)^3$, где $x \in [x_i, x_{i+1}],\ i = \overline{1, n-1}$

Здесь необходимо определить значения коэффициентов $a_i,\ b_i,\ c_i,\ d_i:\ всего\ 4\cdot(n-1)\ переменных$

Обозначим ${m_i}$ значение второй производной в точке ${x_i}$. Тогда для определения коэффициентов сплайна $a_i, b_i, c_i, d_i$ на промежутке $[x_i, x_{i+1}]$ необходимо выполнение следующих требований:

$
\begin{equation}
\left \{ \begin{array}{l}
S_i(x_i)=y_i \\
S_i(x_{i+1}) = y_{i+1} \\
S_i''(x_i) = m_i \\
S_i''(x_{i+1}) = m_{i+1}
\end{array}\right.
\label{eq:ref}
\end{equation}
$
